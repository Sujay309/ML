linear Regression models the relationship between the dependent variable(y) and one or more independent variables (x) by fitting a linear equation to observed data.  in simple terms, it finds the best-fitting straight line through the data points.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the data
# hear we looding a data from sklearn

from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
data = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
```

```python
data['target'] = diabetes.target
```

```python
data
```

```python
# EDA, DATA-CLEANING, DATA-PREPARATION, FEATURE ENGINEERING
```

```python
# DATA DIVIDE THE DATA INTO X, Y
#simple linear regration we take only one feature


x = data[['bmi']]
y = data['target']
```

```python
x
```

```python
y
```

```python
# split the data into tran and test
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# scaling (optional)
```

```python
# model training

from sklearn.linear_model import LinearRegression
```

```python
#instaniate an LR model
model = LinearRegression()
```

```python
model
```

```python
x_train
```

```python
# training of model
model.fit(x_train, y_train)  # at backend, it will use gradient descent to give optimal coefficient
```

```python
#coefficient
model.coef_
```

```python
# intercept
model.intercept_
```

```python
# using x_test data pradict the y_pradict
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
#visualise the result
plt.scatter(x_test, y_test, color = 'black', label = 'Actual data')
plt.plot(x_test, y_pred, color = 'blue', linewidth = 3, label = "Linear regression line")
plt.xlabel("BMI")
plt.ylabel("One year progression-terget")
plt.title("Linear regression on diabetes data")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred)
```
